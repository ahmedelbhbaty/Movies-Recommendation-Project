```python
import ast

def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_ds['genres'] = movies_ds['genres'].apply(convert)
movies_ds['keywords'] = movies_ds['keywords'].apply(convert)
movies_ds.head()
```

```python
import ast

def convert2(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter +=1
        else: 
            break
        return L
```

```python
movies_ds['cast'] = movies_ds['cast'].apply(convert2)
```

```python
movies_ds.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director' :
            L.append(i['name'])
            break
    return L
        
```

```python
movies_ds['crew'] = movies_ds['crew'].apply(fetch_director)
```

```python
movies_ds['overview']= movies_ds['overview'].apply(lambda x:x.split())
```

```python
movies_ds['genres']=movies_ds['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies_ds['Keywords']=movies_ds['keywords'].apply(lambda x:[i.replace(" ","") for i in x])

movies_ds['crew']=movies_ds['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_ds.head(20)
```

```python
# Convert non-list elements to lists

#This code snippet uses the isinstance function to check if each element in the specified columns is a list. If it's not a list, 
#it converts it to a single-element list. After this conversion, you can concatenate the columns without encountering the TypeError.

for column in ['overview', 'budget', 'genres', 'keywords', 'cast', 'crew']:
    movies_ds[column] = movies_ds[column].apply(lambda x: x if isinstance(x, list) else [x])

# Concatenate the columns
movies_ds['tags'] = movies_ds['overview'] + movies_ds['budget'] + movies_ds['genres'] + \
                    movies_ds['keywords'] + movies_ds['cast'] + movies_ds['crew']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features= 5000, stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[0].shape
```

```python
sorted(list(enumerate(similarity[0])),reverse= True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse= True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)        
```
